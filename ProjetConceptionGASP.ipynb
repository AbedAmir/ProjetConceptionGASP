```python
from pyspark import SparkContext, SparkConf
sc = SparkContext.getOrCreate()
```

```python
# Lecture a partir d'un CSV 
data = spark.read.option("header",True).option("inferSchema",True).csv("C:/Users/Winsido/Desktop/VGP-week3-data.csv")
data.printSchema()
data.collect()
#Le resultat est une RDD
```

```python
#Test filter sur une RDD
dataFilter = data.rdd.filter(lambda l : l[3]>48.79)
dataFilter.collect()
```

```python
#Creation de la dataFrame
df = spark.createDataFrame(dataFilter)
df.printSchema()
df.show(10)
```

```python
#Application filtre sur la DF
dfFilter = df.filter(df.kit_id>81)
dfFilter.show()
```
